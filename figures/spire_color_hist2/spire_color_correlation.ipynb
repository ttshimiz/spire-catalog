```python
#Upload the HRS data
#We apply corrections to the fluxes to account for the different beam areas used in the photometry of the BAT AGN and HRS
#250 micron: 423^2/465^2
#350 micron: 751^2/822^2
#500 micron: 1587^2/1769^2
dir = '/Users/ttshimiz/Dropbox/Research/Thesis/'
hrs_data = np.genfromtxt(dir+'Other_surveys/HRS/HRS_PHOTOMETRY_v2.1.txt', dtype = None, names = True)
hrs_250 = hrs_data['S250']*423.**2/465.**2
hrs_250_err = hrs_data['err_tot250']*423.**2/465.**2
hrs_350 = hrs_data['S350']*751.**2/822.**2
hrs_350_err = hrs_data['err_tot350']*751.**2/822.**2
hrs_500 = hrs_data['S500']*1587.**2/1769.**2
hrs_500_err = hrs_data['err_tot500']*1587.**2/1769.**2
hrs_names = hrs_data['HRS_1']
hrs_250_flg = hrs_data['Flag250']
hrs_350_flg = hrs_data['Flag350']
hrs_500_flg = hrs_data['Flag500']
```

```python
#Only use sources that are detected and don't have possible contamination
#These are indicated as having 'flag' = 1 or 2
ind_hrs = (((hrs_250_flg == 1) | (hrs_250_flg == 2)) &
           ((hrs_350_flg == 1) | (hrs_350_flg == 2)) &
           ((hrs_500_flg == 1) | (hrs_500_flg == 2)))
```

```python
#Upload the BAT data
#SPIRE fluxes are contained in the variables: h250, h350, h500 with the errors in h250_err, h350_err, h500_err
#Only use the sources which are detected in all 3 bands
execfile(dir+'scripts/upload_bat_ir_database.py')
ind_bat = (h250 != 0) & (h350 != 0) & (h500 != 0)
```

```python
#Calculate colors
hrs_color_250_350 = hrs_250[ind_hrs]/hrs_350[ind_hrs]
hrs_color_350_500 = hrs_350[ind_hrs]/hrs_500[ind_hrs]
bat_color_250_350 = h250[ind_bat]/h350[ind_bat]
bat_color_350_500 = h350[ind_bat]/h500[ind_bat]
```

```python
#Calculate the theoretical colors based on a simple modified blackbody (greybody) with S = v^b * B(v,T)
#b = dust emissivity, T = dust temperature, v = frequency, and B(v,T) is the blackbody function
beta1 = 2.0
beta2 = 1.5
temp = np.arange(10, 60.1, 0.1)
nu_250 = 3e10/(250*10**(-4))
nu_350 = 3e10/(350*10**(-4))
nu_500 = 3e10/(500*10**(-4))
planckh = 6.6262e-27
boltzmannk = 1.381e-16

theory_color_250_350_1 = ((nu_250/nu_350)**(beta1)*(nu_250/nu_350)**(3)*(np.exp(planckh*nu_350/boltzmannk/temp)-1) /
                          (np.exp(planckh*nu_250/boltzmannk/temp)-1))
theory_color_250_350_2 = ((nu_250/nu_350)**(beta2)*(nu_250/nu_350)**(3)*(np.exp(planckh*nu_350/boltzmannk/temp)-1) /
                          (np.exp(planckh*nu_250/boltzmannk/temp)-1))
theory_color_350_500_1 = ((nu_350/nu_500)**(beta1)*(nu_350/nu_500)**(3)*(np.exp(planckh*nu_500/boltzmannk/temp)-1) /
                          (np.exp(planckh*nu_350/boltzmannk/temp)-1))
theory_color_350_500_2 = ((nu_350/nu_500)**(beta2)*(nu_350/nu_500)**(3)*(np.exp(planckh*nu_500/boltzmannk/temp)-1) /
                          (np.exp(planckh*nu_350/boltzmannk/temp)-1))
```

```python
#Plot the colors both observed and theoretical
clf()
#fig = plt.figure(figsize=(10, 7.5))
fig = plt.figure(figsize=(fw_column*1.33, fw_column))
ax = fig.add_subplot(111)

ax.plot(hrs_color_350_500, hrs_color_250_350, label='HRS', ls='None', marker='o', mec='white', alpha=0.75, ms=4)
ax.plot(bat_color_350_500, bat_color_250_350, label='BAT', ls='None', marker='^', mec='white')
#ax.plot(theory_color_350_500_1, theory_color_250_350_1, 'm-', lw=1.5, label='_nolabel')
#ax.plot(theory_color_350_500_2, theory_color_250_350_2, ls='-', lw=1.5, color='gold', label='_nolabel')
ax.plot(theory_color_350_500_1[range(0,550, 50)], theory_color_250_350_1[range(0,550, 50)], marker='s',
        label=r'$\beta=2.0$', mec='white')
ax.plot(theory_color_350_500_2[range(0,550, 50)], theory_color_250_350_2[range(0,550, 50)], marker='d',
        label=r'$\beta=1.5$', mec='white')
ax.legend(loc='upper left')
ax.set_xlim(0.5, 4.0)
ax.set_ylim(0.5, 3.5)
ax.set_xlabel(r'$F_{350}/F_{500}$')
ax.set_ylabel(r'$F_{250}/F_{350}$')
#plt.show()
```
