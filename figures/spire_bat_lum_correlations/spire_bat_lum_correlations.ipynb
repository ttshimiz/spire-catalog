# BAT Luminosity vs. SPIRE Luminosities

In this notebook we calculate the SPIRE and BAT luminosities and plot them against each other. We split the sample into Seyfert 1's and 2's and will also plot the upper limits as downward-facing grey arrows. 
The final figure will consist of 6 separate plots in a 2x3 array. The top row will be for Sy 1's and the bottom row for Sy 2's. Each column will be for SPIRE 250, 350, and 500 micron.

```python
import matplotlib.pyplot as plt
```

```python
#Plotting parameters
marker_size = 4          # Size of the markers in the plots
arrow_width = 0.007      # Relative width of the upper limit arrows
arrow_length = 4.5*2.25  # Length of the arrows
columnwidth = 245.26653   #in points
textwidth = 513.11743     #in points
pts2inches = 1./72.27    #Conversion from pts to inches
fw_column = columnwidth * pts2inches  #Figure width for a figure that spans the width of a column
fw_text   = textwidth * pts2inches    #Figure width for a figure that spans the width of the page
```

```python
#Upload all of the data. In particular this script uploads the SPIRE fluxes and BAT fluxes and converts them to luminosities.
#It also separates the sample into Sy 1's and Sy 2's.
#The relevant variables that it creates are l250, l350, l500, lbat, sy1_sample, and sy2_sample.
#sy1_sample is a boolean array that will filter out all of the Sy 1's. Likewise for sy2_sample and Sy 2's.

dir = '/Users/ttshimiz/Dropbox/Research/Thesis/'
execfile(dir+'scripts/upload_bat_ir_database.py')
```

```python
#Pull out all of the undetected galaxies and separate into Sy 1 and Sy 2
undetected250 = l250 == 0                 #Undetected points have a flux of zero
undetected350 = l350 == 0
undetected500 = l500 == 0

l250_sy1_undetected = l250_err[undetected250 & sy1_sample] #The 5sigma upper limit is then the error
l250_sy2_undetected = l250_err[undetected250 & sy2_sample]
l350_sy1_undetected = l350_err[undetected350 & sy1_sample] 
l350_sy2_undetected = l350_err[undetected350 & sy2_sample]
l500_sy1_undetected = l500_err[undetected500 & sy1_sample] 
l500_sy2_undetected = l500_err[undetected500 & sy2_sample]

l250_sy1_detected = l250[~undetected250 & sy1_sample]
l250_sy2_detected = l250[~undetected250 & sy2_sample]
l350_sy1_detected = l350[~undetected350 & sy1_sample]
l350_sy2_detected = l350[~undetected350 & sy2_sample]
l500_sy1_detected = l500[~undetected500 & sy1_sample]
l500_sy2_detected = l500[~undetected500 & sy2_sample]
```

```python
#Setup the figure
fig = plt.figure(figsize=(fw_text, fw_text*2./3.))

#L250 vs. LBAT for Sy 1
ax1 = fig.add_subplot(2,3,1)

#Plot the undetected points with arrows using quiver
ax1.quiver(lbat[undetected250 & sy1_sample], l250_sy1_undetected, 
           zeros(len(l250_sy1_undetected)), -ones(len(l250_sy1_undetected)),
           width = arrow_width, scale = arrow_length, scale_units='height', color = 'gray')

#Plot the detected points with open circles
ax1.plot(lbat[~undetected250 & sy1_sample], l250_sy1_detected,  'bo', ms=marker_size,
         fillstyle='none', mew=0.75)

ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_ylabel(r'$L_{\rm{BAT}}$ [ergs s$^{-1}$]', fontsize = 10)
ax1.set_ylim(1e40, 3e44)
ax1.set_xlim(1e41, 1e45)
ax1.tick_params(axis='both', which='both', labelbottom='off', labelsize=10)
ax1.text(1.5e41, 2e44, 'Sy 1', fontsize=10, ha='left', va='top')

#L350 vs. LBAT for Sy 1
ax2 = fig.add_subplot(2,3,2)

#Plot the undetected points with arrows using quiver
ax2.quiver(lbat[undetected350 & sy1_sample], l350_sy1_undetected, 
           zeros(len(l350_sy1_undetected)), -ones(len(l350_sy1_undetected)),
           width = arrow_width, scale = arrow_length, scale_units='height', color = 'gray')

#Plot the detected points with open circles
ax2.plot(lbat[~undetected350 & sy1_sample], l350_sy1_detected,  'bo', ms=marker_size,
         fillstyle='none', mew=0.75)

ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_ylim(1e40, 3e44)
ax2.set_xlim(1e41, 1e45)
ax2.tick_params(axis='both', which='both', labelbottom='off', labelleft='off')

#L500 vs. LBAT for Sy 1
ax3 = fig.add_subplot(2,3,3)

#Plot the undetected points with arrows using quiver
ax3.quiver(lbat[undetected500 & sy1_sample], l500_sy1_undetected, 
           zeros(len(l500_sy1_undetected)), -ones(len(l500_sy1_undetected)),
           width = arrow_width, scale = arrow_length, scale_units='height', color = 'gray')

#Plot the detected points with open circles
ax3.plot(lbat[~undetected500 & sy1_sample], l500_sy1_detected,  'bo', ms=marker_size,
         fillstyle='none', mew=0.75)

ax3.set_xscale('log')
ax3.set_yscale('log')
ax3.set_ylim(1e40, 3e44)
ax3.set_xlim(1e41, 1e45)
ax3.tick_params(axis='both', which='both', labelbottom='off', labelleft='off')

#L250 vs. LBAT for Sy 2
ax4 = fig.add_subplot(2,3,4)

#Plot the undetected points with arrows using quiver
ax4.quiver(lbat[undetected250 & sy2_sample], l250_sy2_undetected, 
           zeros(len(l250_sy2_undetected)), -ones(len(l250_sy2_undetected)),
           width = arrow_width, scale = arrow_length, scale_units='height', color = 'gray')

#Plot the detected points with open circles
ax4.plot(lbat[~undetected250 & sy2_sample], l250_sy2_detected,  'rs', ms=marker_size,
         fillstyle='none', mew=0.75)

ax4.set_xscale('log')
ax4.set_yscale('log')
ax4.set_xlabel(r'$L_{\rm{250}}$ [ergs s$^{-1}$]', fontsize = 10)
ax4.set_ylabel(r'$L_{\rm{BAT}}$ [ergs s$^{-1}$]', fontsize = 10)
ax4.set_ylim(1e40, 3e44)
ax4.set_xlim(1e41, 1e45)
ax4.tick_params(axis='both', which='major', labelsize=10)
ax4.text(1.5e41, 2e44, 'Sy 2', fontsize=10, ha='left', va='top')

#L350 vs. LBAT for Sy 2
ax5 = fig.add_subplot(2,3,5)

#Plot the undetected points with arrows using quiver
ax5.quiver(lbat[undetected350 & sy2_sample], l350_sy2_undetected, 
           zeros(len(l350_sy2_undetected)), -ones(len(l350_sy2_undetected)),
           width = arrow_width, scale = arrow_length, scale_units='height', color = 'gray')

#Plot the detected points with open circles
ax5.plot(lbat[~undetected350 & sy2_sample], l350_sy2_detected,  'rs', ms=marker_size,
         fillstyle='none', mew=0.75)

ax5.set_xscale('log')
ax5.set_yscale('log')
ax5.set_xlabel(r'$L_{\rm{350}}$ [ergs s$^{-1}$]', fontsize = 10)
ax5.set_ylim(1e40, 3e44)
ax5.set_xlim(1e41, 1e45)
ax5.tick_params(axis='both', which='both', labelleft='off', labelsize=10)

#L500 vs. LBAT for Sy 2
ax6 = fig.add_subplot(2,3,6)

#Plot the undetected points with arrows using quiver
ax6.quiver(lbat[undetected500 & sy2_sample],l500_sy2_undetected, 
           zeros(len(l500_sy2_undetected)), -ones(len(l500_sy2_undetected)),
           width = arrow_width, scale = arrow_length, scale_units='height', color = 'gray')

#Plot the detected points with open circles
ax6.plot(lbat[~undetected500 & sy2_sample], l500_sy2_detected,  'rs', ms=marker_size,
         fillstyle='none', mew=0.75)

ax6.set_xscale('log')
ax6.set_yscale('log')
ax6.set_xlabel(r'$L_{\rm{500}}$ [ergs s$^{-1}$]', fontsize = 10)
ax6.set_ylim(1e40, 3e44)
ax6.set_xlim(1e41, 1e45)
ax6.tick_params(axis='both', which='major', labelsize=10, labelleft='off')

plt.show()

#Save the figure
fig.savefig(dir+'SPIRE_photometry/Paper/Figures/spire_bat_lum_correlations.eps', bbox_inches='tight')
```
